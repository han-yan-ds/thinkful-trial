```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rentDf = pd.read_excel('Unit2-WhatDoYouSee/fy2019_safmrs.xlsx')
rentDf.head()
```

```python
list(rentDf.columns)
```

```python
rentDf.describe()
```

```python
#I want to make the names simpler

myList = [item.replace('\n','-') for item in rentDf.columns]
myList = [item.replace('SAFMR-','') for item in myList]
myList = [item.replace('HUD ','') for item in myList]
myList = [item.replace(' --90%-Payment-Standard','-Low-End') for item in myList]
myList = [item.replace(' --110%-Payment-Standard','-High-End') for item in myList]
myList[2] = 'Area-Name'
myList[1] = 'Area-Code'
myList
```

```python
rentDf.columns = myList
rentDf.tail()
```

```python
#test out function for creating new column for State

abileneTxStr = rentDf.loc[0,'Area-Name']
abileneTxStr[abileneTxStr.find(', ')+2:abileneTxStr.find(', ')+4]
```

```python
#Create new column with State

rentDf['State'] = rentDf['Area-Name'].apply(lambda string: string[string.find(', ')+2:string.find(', ')+4])
```





# Plot 1: Colorado Rents by Metro Area

```python
boxplot_CO1bd = sns.boxplot(x='Area-Name', y='1BR', data=rentDfCO)
boxplot_CO1bd.set_xticklabels(rotation=90, labels=boxplot_CO1bd.get_xticklabels())
boxplot_CO1bd.set_ylabel('1-BR Rental Prices')
boxplot_CO1bd.set_title('Rental Prices (Colorado, 1 Bedroom)', fontdict={'weight':'bold'})
```

### We can clearly see that Denver and Boulder are the 2 main Metro areas with the highest rent.

### Denver's rent covers the largest range, which makes sense because it includes Lakewood, Aurora, and downtown Denver

### The only Metro area in this data set NOT in the Front Range is Grand Junction, which explains why the rent is so cheap there.

### I always wondered if there's a trend that Zip Codes go up as you go west.  This is True!

# Plot 4: How do rents compare in different areas in California, and with different sizes of units?

```python
rentDfCA = rentDf[ rentDf['State']=='CA' ]
rentDfCA['Area-Name'].unique()
rentDfCA.head()
```

DataFrame.melt() UNPIVOTS a table

```python
rentDfCA = rentDfCA.melt(id_vars=['ZIP-Code', 'Area-Name'], value_vars=['0BR', '1BR', '2BR', '3BR'])
rentDfCA.columns = ['ZIP-Code', 'Area-Name', 'Housing-Size', 'Rent']
```

```python
rentDfCA['Area-Name'] = rentDfCA['Area-Name'].apply(lambda longString: longString[:longString.find(', CA')])
```

```python
sortedRentDfCA = rentDfCA.sort_values('Rent', ascending=False)
sortedRentDfCA.head()
```

```python
fig2 = plt.Figure()
fig2, axe2 = plt.subplots(figsize=(20,10))

caliRentPlots = sns.barplot(x='Area-Name', y='Rent', hue='Housing-Size', data=sortedRentDfCA)
caliRentPlots.set_xticklabels(rotation=90, labels=caliRentPlots.get_xticklabels())
caliRentPlots.set_title('Rents in California Metros', fontdict={'weight':'bold', 'size':20})

plt.show()
```

### No surprise, the Bay Area is the most expensive.  Interestingly however, San Diego is more expensive than LA

### There seems to be a HUGE price jump from 2BR to 3BR.  The jump from 0BR to 1BR is the smallest by far (it's studio to 1BR which is minor).
