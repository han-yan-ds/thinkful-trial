```python
import numpy as np
import matplotlib.pyplot as plt
```

# Part 1: Other Distribution Types

### 1) Poisson Distribution

```python
#Generating a Poisson distribution

poissonVar = np.random.poisson(3, 100)

plt.hist(poissonVar, color='orange', bins=25)

poissonMean = poissonVar.mean()
poissonStd = poissonVar.std(ddof=0)

plt.axvline(poissonMean, linestyle='solid')
plt.axvline(poissonMean+poissonStd, linestyle='dashed', color='b')
plt.axvline(poissonMean-poissonStd, linestyle='dashed', color='b')
plt.title('Poisson Dist, 100 population, expected freq=3')
plt.ylabel('# Occurrences')

plt.show()
```

```python
#Mean can be useful, but standard deviation is not.  Standard deviation is only useful for a measurement of symmetrical variance, but Poisson is skewed.
```

### 2) Pareto Distribution

```python
#Generating a Pareto distribution

paretoVar = np.random.pareto(1.16, 100)

plt.hist(paretoVar, color='orange', bins=25)

paretoMean = paretoVar.mean()
paretoStd = paretoVar.std(ddof=0)

plt.axvline(paretoMean, linestyle='solid')
plt.axvline(paretoMean+paretoStd, linestyle='dashed', color='b')
plt.axvline(paretoMean-paretoStd, linestyle='dashed', color='b')
plt.title('Pareto Dist, 100 population, 80/20 parameters')

plt.show()
```

```python
#Mean is not useful, because the Pareto distribution has a lot of outliers.  Use Median or Mode instead
#Standard deviation is not useful because of the outliers
```

### 3) Von Mises Distribution

```python
#Generating a von Mises distribution (normal dist on a unit circle from -pi to pi)

vonmisesVar = np.random.vonmises(2, 1, 100)

plt.hist(vonmisesVar, color='orange', bins=25)

vonmisesMean = vonmisesVar.mean()
vonmisesStd = vonmisesVar.std(ddof=0)

plt.axvline(vonmisesMean, linestyle='solid')
plt.axvline(vonmisesMean+vonmisesStd, linestyle='dashed', color='b')
plt.axvline(vonmisesMean-vonmisesStd, linestyle='dashed', color='b')
plt.title('von Mises Dist, 100 population, mode=2, dispersion=1')

plt.show()
```

```python
#Mean is useful (because this is essentially a normal dist on a circle), but keep in mind the numbers outside of (-pi, pi) "wraps", much like angles outside (0, 360)
#Standard deviation works, but again, beware of the "wrap"
```

### 4) Exponential Distribution

```python
#Generating an Exponential distribution

exponentialVar = np.random.exponential(2, 100)

plt.hist(exponentialVar, color='orange', bins=25)

exponentialMean = exponentialVar.mean()
exponentialStd = exponentialVar.std(ddof=0)

plt.axvline(exponentialMean, linestyle='solid')
plt.axvline(exponentialMean+exponentialStd, linestyle='dashed', color='b')
plt.axvline(exponentialMean-exponentialStd, linestyle='dashed', color='b')
plt.title('Exponential Dist, 100 population, lambda=1/2')

plt.show()
```

```python
#Mean is not useful because data is very skewed, and there's outliers.  Use median or mode.
#Standard deviation is not useful because of the skewness
```

### 5) Binomial Distribution

```python
#Generating a Binomial distribution

binomialVar = np.random.binomial(30, 0.25, 100)

plt.hist(binomialVar, color='orange', bins=25)

binomialMean = binomialVar.mean()
binomialStd = binomialVar.std(ddof=0)

plt.axvline(binomialMean, linestyle='solid')
plt.axvline(binomialMean+binomialStd, linestyle='dashed', color='b')
plt.axvline(binomialMean-binomialStd, linestyle='dashed', color='b')
plt.title('Binomial Dist, 100 population, 30 trials, p=0.25 per trial')
plt.ylabel('# Successes Given 30 Trials')

plt.show()
```

```python
#Mean makes sense in Binomial, because there's a similar chance in either direction of the expected value (n*p)
#Standard deviations, at first glance, makes sense, because the histogram has a bell shape, and a good estimate of around 68% fit within 1 standard deviation of the mean
```

### 6) Negative Binomial Distribution

```python
#Generating a Negative Binomial distribution

negbinomialVar = np.random.negative_binomial(10, 0.25, 100)

plt.hist(negbinomialVar, color='orange', bins=25)

negbinomialMean = negbinomialVar.mean()
negbinomialStd = negbinomialVar.std(ddof=0)

plt.axvline(negbinomialMean, linestyle='solid')
plt.axvline(negbinomialMean+negbinomialStd, linestyle='dashed', color='b')
plt.axvline(negbinomialMean-negbinomialStd, linestyle='dashed', color='b')
plt.title('Gamma Dist, 100 population, 10 successes, p=0.25 per trial')
plt.ylabel('# of Failures Given 10 Successes')

plt.show()
```

```python
#Mean makes sense in Negative Binomial, because there's a similar chance in either direction of the expected value (n/(1-p))
#Standard deviations, at first glance, makes sense, because the histogram has a bell shape, and a good estimate of around 68% fit within 1 standard deviation of the mean
```

# Part 2: Adding normal distributions

```python
norm1 = np.random.normal(5, 0.5, 1000)
norm2 = np.random.normal(10, 1, 1000)
```

```python
normSum = norm1 + norm2
```

```python
plt.hist(normSum, color='pink')
plt.axvline(normSum.mean()+normSum.std(ddof=0), linestyle='dashed', color='grey')
plt.axvline(normSum.mean()-normSum.std(ddof=0), linestyle='dashed', color='grey')
plt.axvline(normSum.mean(), linestyle='solid', color='black')

plt.show()
```

```python
#If I add 2 normal distributions, the mean is the sum of the parts.
```

```python
print(norm1.std(ddof=0))
print(norm2.std(ddof=0))
print(normSum.std(ddof=0))
```

```python
#If I add 2 normal distributions, the variance is the sum of the variances.  
#So the standard deviation is the squareroot of the sum of the squares
```
